```python
import pandas as pd
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split
import tensorflow as tf
import matplotlib.pyplot as plt

# load CSV file
df = pd.read_csv("C:\\Users\\Shubham\\Desktop\\New Merge leader\\Amit Shah merge.csv")

# drop rows with non-string values in the "Text" column
df = df.dropna(subset=["Text"])
df = df[df["Text"].apply(lambda x: isinstance(x, str))]

# preprocess text
tokenizer = Tokenizer(num_words=10000, oov_token="<OOV>")
tokenizer.fit_on_texts(df["Text"])
sequences = tokenizer.texts_to_sequences(df["Text"])
padded_sequences = pad_sequences(sequences, maxlen=50, padding="post", truncating="post")

# split data into training and testing sets
labels = pd.get_dummies(df["score"], dtype=np.float32).values
X_train, X_test, y_train, y_test = train_test_split(padded_sequences, labels, test_size=0.2, random_state=42)

# build RNN model
model = tf.keras.Sequential([
    tf.keras.layers.Embedding(10000, 16, input_length=50),
    tf.keras.layers.LSTM(16),
    tf.keras.layers.Dense(3, activation="softmax")
])
model.compile(loss="categorical_crossentropy", optimizer="adam", metrics=["accuracy"])

# train model
model.fit(X_train, y_train, epochs=5, batch_size=32, validation_data=(X_test, y_test))

# predict sentiment scores for test data
y_pred = model.predict(X_test)

# count number of positive, negative, and neutral predictions
positive_count = 0
negative_count = 0
neutral_count = 0
for i in range(len(y_pred)):
    if np.argmax(y_pred[i]) == 0:
        negative_count += 1
    elif np.argmax(y_pred[i]) == 1:
        neutral_count += 1
    elif np.argmax(y_pred[i]) == 2:
        positive_count += 1

# plot pie chart
labels = ['Positive', 'Negative', 'Neutral']
sizes = [positive_count, negative_count, neutral_count]
colors = ['green', 'red', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Amit Shah Pie Chart')

plt.show()
```
