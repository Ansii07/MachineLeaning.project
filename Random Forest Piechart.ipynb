```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
import pandas as pd

# Read the CSV file
data = pd.read_csv("C:\\Users\\Shubham\\Desktop\\New Merge leader\\Amit Shah merge.csv")

# Extract the text data and corresponding labels
text_data = data['Text']  # Replace 'text_column' with the actual column name containing the text
labels = data['score']  # Replace 'target_variable' with the actual column name of the target variable

# Vectorize the text data using TF-IDF
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(text_data)

# Initialize and train the Random Forest model
rf_model = RandomForestClassifier()
rf_model.fit(X, labels)

# Make predictions using the trained model
predictions = rf_model.predict(X)

# Calculate the count of each unique prediction
prediction_counts = pd.Series(predictions).value_counts()

# Define colors for each prediction category
colors = ['green', 'yellow', 'red']  # Modify colors as needed

# Create a pie chart with color coding
plt.pie(prediction_counts, labels=prediction_counts.index, colors=colors, autopct='%1.1f%%')
# plt.title('Amit Shah Pie Chart')
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.show()
```
