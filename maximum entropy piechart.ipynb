```python
import pandas as pd
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt

import nltk
nltk.download('vader_lexicon')


# Load the CSV file into a pandas dataframe
df = pd.read_csv("C:\\Users\\Shubham\\Desktop\\New Merge leader\\Amit Shah merge.csv")

# Create a SentimentIntensityAnalyzer object
sia = SentimentIntensityAnalyzer()

# Use NLTK to perform sentiment analysis on the text data in the dataframe
sentiments = []
for text in df['Text']:
    score = sia.polarity_scores(text)
    if score['compound'] >= 0.05:
        sentiments.append('positive')
    elif score['compound'] <= -0.05:
        sentiments.append('negative')
    else:
        sentiments.append('neutral')

# Count the number of positive, negative, and neutral sentiments
counts = pd.Series(sentiments).value_counts()

# Create a pie chart showing the distribution of sentiments
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('Amit saha Maximum Entropy')
plt.show()
```
