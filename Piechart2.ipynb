```python
#x

import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load the CSV merge file into a pandas dataframe
df = pd.read_csv("C:\\Users\\Shubham\\Desktop\\New Merge leader\\Yogi merge.csv")

# Clean the text data and convert it to string format
df['score'] = df['score'].astype(str)

# Perform sentiment analysis on each entry in the file
sentiments = []
for entry in df['score']:
    blob = TextBlob(entry)
    sentiment_score = blob.sentiment.polarity
    if sentiment_score > 0:
        sentiments.append('positive')
    elif sentiment_score < 0:
        sentiments.append('negative')
    else:
        sentiments.append('neutral')

# Count the number of entries in each sentiment category
counts = pd.Series(sentiments).value_counts()

# Create a pie chart to visualize the distribution of sentiment categories
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title('All factor of yogi')
plt.show()

```

```python
#x

import pandas as pd
from textblob import TextBlob
import matplotlib.pyplot as plt

# Load the CSV merge file into a pandas dataframe
df = pd.read_csv("C:\\Users\\Shubham\\Desktop\\Leader Merge File.csv")

# Define the factors to analyze
factors = ['amit', 'Arvind Kejriwal', 'Mamata Banerjee', 'Narendra Modi', 'Rahul Gandhi', 'rajnath']

# Clean the text data and convert it to string format
for score in factors:
    df['score'] = df['score'].apply(lambda x: str(x))

# Perform sentiment analysis on each entry in the file for each factor
sentiments = {}
for factor in factors:
    factor_sentiments = []
    for entry in df['score']:
        blob = TextBlob(entry)
        sentiment_score = blob.sentiment.polarity
        if sentiment_score > 0:
            factor_sentiments.append('positive')
        elif sentiment_score < 0:
            factor_sentiments.append('negative')
        else:
            factor_sentiments.append('neutral')
    sentiments[factor] = pd.Series(factor_sentiments).value_counts()

# Plot the sentiment analysis results for each factor in a single pie chart
fig, axs = plt.subplots(2, 3, figsize=(10, 6))
axs = axs.flatten()
for i, factor in enumerate(factors):
    axs[i].pie(sentiments[factor], labels=sentiments[factor].index, autopct='%1.1f%%')
    axs[i].set_title(factor)
plt.suptitle('Sentiment Analysis by Factor')
plt.show()
```
